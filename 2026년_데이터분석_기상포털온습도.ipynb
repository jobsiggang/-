## 데이터 전처리 및 병합


```
#데이터 병합
df_merge = pd.merge(df_temp, df_humi, on=["지점번호", "지점명", "일시"], how="inner")
# df_merge = pd.merge(df_temp, df_humi, on=[ "일시"], how="inner") #이 코드로 했을때 문제점을 비교해 보세요.
df_merge.describe()
```

```
# 일시 항목의 형식을 문자열에서 날짜 형식으로 변환
df_merge['일시']=pd.to_datetime(df_merge['일시'])
```

```
# ‘연도’와 ‘월’ 항목을 추가
df_merge["연도"] = df_merge["일시"].dt.year
df_merge["월"] = df_merge["일시"].dt.month
```

# 데이터의 정규화

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# 1. 정규화할 컬럼 선택
cols = ['평균기온(℃)', '평균습도(%rh)']

# 2. 최소-최대 정규화 (Min-Max Scaling) 수행
# 데이터를 0과 1 사이의 값으로 변환합니다.
scaler = MinMaxScaler()
df_scaled = df_merge.copy()
df_scaled[cols] = scaler.fit_transform(df_merge[cols])

# 3. 박스플롯 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 정규화 전 (Raw Data)
sns.boxplot(data=df_merge[cols], ax=axes[0])
axes[0].set_title('정규화 전 (원본 데이터)')
axes[0].set_ylabel('값의 범위')

# 정규화 후 (Min-Max Scaled)
sns.boxplot(data=df_scaled[cols], ax=axes[1])
axes[1].set_title('정규화 후 (0~1 범위로 변환)')
axes[1].set_ylabel('조정된 값 (0~1)')

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import platform


# ---------------------------------------------------------
# [1] 데이터 전처리
# ---------------------------------------------------------
# 월별로 그룹화하여 평균값 계산
# (가정: df_merge에 '월' 컬럼과 수치 데이터가 있다고 가정)
df_grouped = df_merge.groupby('월')[['평균기온(℃)', '평균습도(%rh)']].mean()

# 정규화할 컬럼 지정
cols = ['평균기온(℃)', '평균습도(%rh)']

# ---------------------------------------------------------
# [2] 최소-최대 정규화 (Min-Max Scaling)
# ---------------------------------------------------------
scaler = MinMaxScaler()

# 원본 데이터 보존을 위해 copy() 사용
df_scaled = df_grouped.copy()

# 정규화 수행 (fit_transform) 및 데이터 교체
df_scaled[cols] = scaler.fit_transform(df_grouped[cols])

# ---------------------------------------------------------
# [3] 꺾은선 그래프 시각화
# ---------------------------------------------------------
plt.figure(figsize=(10, 6)) # 그래프 크기 설정 (가로, 세로)

# 기온 그래프 그리기 (빨간색, 마커는 동그라미, 라벨 추가)
plt.plot(df_scaled.index, df_scaled['평균기온(℃)'],
         color='red', marker='o', label='평균기온')

# 습도 그래프 그리기 (파란색, 마커는 네모, 라벨 추가)
plt.plot(df_scaled.index, df_scaled['평균습도(%rh)'],
         color='blue', marker='s', label='평균습도')

# 그래프 꾸미기
plt.title('월별 평균 기온 및 습도 변화', fontsize=15) # 제목
plt.xlabel('월') # X축 라벨
plt.ylabel('조정된 값 (0 ~ 1)') # Y축 라벨
plt.legend() # 범례 표시 (label 인자 활용)
plt.grid(True, linestyle='--', alpha=0.6) # 격자 표시 (점선, 투명도 조절)

plt.tight_layout() # 여백 자동 조정
plt.show()
```

```
# x축: 평균기온, y축: 평균습도
# regplot은 산점도와 회귀선(추세선)을 같이 그려줍니다.
sns.regplot(data=df_merge, x='평균기온(℃)', y='평균습도(%rh)')
plt.title('기온과 습도의 상관관계 분석')
plt.show()
```

```
# 수치형 컬럼만 선택
cols = ["평균기온(℃)", "최고기온(℃)", "최저기온(℃)", "일교차",
        "평균습도(%rh)", "최저습도(%rh)"]
corr = df_merge[cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("울산 기온·습도 변수 간 상관관계")
plt.show()
```
